```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le=LabelEncoder()
df['species']=le.fit_transform(df['species'])
```

```python
df.to_csv('iris_cleaned.csv',index=False)
```

```python
df['petal_length'].hist(bins=50)
plt.suptitle("petal_length")
plt.show()
```

```python
df.hist(bins=10,figsize=(8,8))
plt.show()
```

```python
sns.countplot(x='species', data=df, palette=['red', 'pink','orange'])
plt.title("species count plot",weight='bold')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="crest" ,linewidths=.5)
```

```python
sns.lineplot(x="petal_width",y="petal_length",data=df)
```

```python
sns.scatterplot(data=df,x="petal_length", y="sepal_length",hue="species")
plt.suptitle("relation b/w sepal length and petal length ")
plt.show()
```

```python
plt.style.use('dark_background')
sns.relplot(x="petal_width", y="species", hue="species",data=df)
```

```python
sns.set_style('whitegrid')
sns.pairplot(df,hue="species",palette=['orange','green','blue'])
```
